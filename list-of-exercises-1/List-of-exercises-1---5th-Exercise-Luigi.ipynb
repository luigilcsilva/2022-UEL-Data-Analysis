First of all, let us import some useful packages.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import os
import scipy.integrate as integrate
import scipy.special as special
import scipy.interpolate as interpolate
import scipy.stats as stats
```

Now, let us generate random numbers for the binomial and the poisson distribution, keeping $\lambda = n p$ constant.

```python
### Quantity of numbers to be generated.
N_all = np.arange(5000, 65000, 5000)

### Parameters for the plot.
col_quantity = 3

if len(N_all)//col_quantity == len(N_all)/col_quantity:
    lines_quantity = len(N_all)//col_quantity
else:
    lines_quantity = len(N_all)//col_quantity + 1
    
size_param = 15/col_quantity
figure, axis = plt.subplots(lines_quantity, col_quantity,
                            figsize=(size_param*col_quantity,size_param*lines_quantity))
bins_def = np.arange(0, 45, 2.5)

### Generating the numbers and plotting them.
j=0
for N in N_all:
    ### Binomial Distribution
    #Seed
    np.random.seed(seed=j)
    
    lamb = 15
    p = lamb/N

    x_binomial = np.random.binomial(N, p, size=N)
    
    ### Poisson distribution
    #Seed
    np.random.seed(seed=j+1)
    
    x_poisson = np.random.poisson(lam=lamb, size=N)
    
    ### Plotting graphs
    ### FIRST PLOT ###
    coord_1 = j//col_quantity
    coord_2 = j%col_quantity
    axis[coord_1, coord_2].hist(x_binomial, alpha=0.7, label='Binomial', bins=bins_def)
    axis[coord_1, coord_2].hist(x_poisson, alpha=0.5, label='Poisson', bins=bins_def)
    axis[coord_1, coord_2].set_xlabel('Value')
    axis[coord_1, coord_2].set_ylabel('Frequency')
    axis[coord_1, coord_2].set_title('Histogram for N='+str(N))
    axis[coord_1, coord_2].legend()
    
    j+=1
```

As we can see above, for $N \approx 50000$, the differences between the binomial and poisson distribution become very small.
