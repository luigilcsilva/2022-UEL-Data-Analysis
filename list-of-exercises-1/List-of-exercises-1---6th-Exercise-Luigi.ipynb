First of all, let us import some useful packages.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import os
import scipy.integrate as integrate
import scipy.special as special
import scipy.interpolate as interpolate
import scipy.stats as stats
```

Now, let us generate random numbers for the poisson distribution with a certain $\lambda$ and for the gaussian distribution with mean $\lambda$ and standard deviation $\sqrt{\lambda}$.

```python
### Defining the lambda values and the quantity of numbers to be generated.
lamb_values = [1, 5, 20, 100, 1000]

N = 100000

### Parameters for the plot.
col_quantity = 3

if len(lamb_values)//col_quantity == len(lamb_values)/col_quantity:
    lines_quantity = len(lamb_values)//col_quantity
else:
    lines_quantity = len(lamb_values)//col_quantity + 1
    
size_param = 15/col_quantity
figure, axis = plt.subplots(lines_quantity, col_quantity,
                            figsize=(size_param*col_quantity,size_param*lines_quantity))

j=0
for lamb in lamb_values:
    #Seed
    np.random.seed(seed=j)
    #Random Numbers
    std_dev = np.sqrt(lamb)
    x_gauss = np.random.normal(loc=lamb, scale=std_dev, size=N)
    
    #Seed
    np.random.seed(seed=j+1)
    #Random Numbers
    x_poisson = np.random.poisson(lam=lamb, size=N)
    
    ### Plotting graphs
    n_bins = 15
    bins_def = np.arange(lamb-3*std_dev, lamb+3*std_dev, 6*std_dev/n_bins)
    
    coord_1 = j//col_quantity
    coord_2 = j%col_quantity
    axis[coord_1, coord_2].hist(x_poisson, alpha=0.7, label='Poisson', bins=bins_def)
    axis[coord_1, coord_2].hist(x_gauss, alpha=0.5, label='Gaussian', bins=bins_def)
    axis[coord_1, coord_2].set_xlabel('Value')
    axis[coord_1, coord_2].set_ylabel('Frequency')
    axis[coord_1, coord_2].set_title('Histogram for $\lambda= %.f$' % lamb)
    axis[coord_1, coord_2].legend()
    
    j+=1
```

As we can see, for $\lambda \gg 1$, the Poisson distribution tends to the Gaussian distribution.
